#### method to parse minutes from datetime data

```python
def get_mins(time_str):
    """Get Seconds from time."""
    h, m, s = time_str.split(':')
    return int(h) * 60 + int(m) + int(s)/60
```

```python
def mins_to_index(minsInterval, listOfMins):
    """
    Turn a minutes time interval into index time interval.
    
    Take as input the minutes time interval, and a list of minutes for the channel in question.
    pc['timeInterval'], loD[0][channelTimevals]
    """
    listOfMins = np.array(listOfMins)
    
    indices = np.where(np.logical_and(listOfMins>=minsInterval['start'], listOfMins<=minsInterval['stop']))
    indices = indices[0]
    
    start = indices[0]
    end = indices[-1]
    
    return start,end
```

#### Processing methods:

```python
def df_to_plotting_dict_Zif268(fdf, blanksDict):
    """
    This method takes the Zif268 platereader data dataframe and generates list of dictionaries
    (for the induced, and uninduced cases) that can be used for plotting (contain the mean signal, st.dev).
    """
    listInduced = []
    listNotInduced = []
    
    for info, strgrp, in fdf.groupby(["Background","Strain"]):
        strain = info[1]
        background = info[0]
        print(background)
        #Plan is to create a list of dictionaries; one dictionary for each curve/dataset in the subsequent plots
        
        #remove nan values, equalize num data points. Calculate  mean, stdev. Pack into dict.
        if background != 'WT':      # The 'WT' background only has uninduced strains
            inducedReplicates = strgrp[strgrp.loc[:,'Media']!='URA'].reset_index()
            dictInduced = process_replicates(strain, background, inducedReplicates, blanksDict)
        
            listInduced.append(dictInduced)
        
        #remove nan values, equalize num data points. Calculate  mean, stdev. Pack into dict.
        notInducedReplicates = strgrp[strgrp.loc[:,'Media']=='URA'].reset_index()
        dictNotInduced = process_replicates(strain, background, notInducedReplicates, blanksDict)
                
        listNotInduced.append(dictNotInduced)
        
    return listInduced, listNotInduced
```

```python
def process_replicates(strain, background, replicategroup, blanksDict):
    """
    This method takes as input a groupby of replicates from the Zif268 DataFrame. It processes 
    this data to remove nan values and equalize the number of data points. It calculates the 
    mean, and StDev. Then it packages everything into a dictionary, which it returns.
    """
    replicateDict = {}
    
    ODarray = []
    GFP70array = []
    GFP100array = []
    SCARLET70array = []
    SCARLET100array = []

    if replicategroup.loc[0,'Media']=='URA':
        induction='uninduced'
    else:
        induction='induced'
    
    if background=='WT':
        mScarlet = False
    else:
        mScarlet = True

    #Combine all of the replicates into a 2D numpy array, remove nan values (found at end)
    for i, replicate in replicategroup.iterrows():            

        ODarray.append([value for value in replicate.loc['OD600'] if str(value) != 'nan'])
        GFP70array.append([value for value in replicate.loc['GFP70'] if str(value) != 'nan'])
        GFP100array.append([value for value in replicate.loc['GFP100'] if str(value) != 'nan'])
        if mScarlet:    
            SCARLET70array.append([value for value in replicate.loc['SCAR70'] if str(value) != 'nan'])
            SCARLET100array.append([value for value in replicate.loc['SCAR100'] if str(value) != 'nan'])

    # Set the number of datapoints in each array to the minimum number of datapoints in any array
    if mScarlet: 
        leastData = min(min([len(array) for array in ODarray]), min([len(array) for array in GFP70array]), min([len(array) for array in GFP100array]), min([len(array) for array in SCARLET70array]),min([len(array) for array in SCARLET100array]))
    else:
        leastData = min(min([len(array) for array in ODarray]), min([len(array) for array in GFP70array]), min([len(array) for array in GFP100array]))

    ODarray = [array[0:leastData] for array in ODarray]
    GFP70array = [array[0:leastData] for array in GFP70array]
    GFP100array = [array[0:leastData] for array in GFP100array]
    if mScarlet: 
        SCARLET70array = [array[0:leastData] for array in SCARLET70array]
        SCARLET100array = [array[0:leastData] for array in SCARLET100array]

    #Get the timeseries' values
    tsOD600 = replicategroup.loc[0,'ts_OD600']
    tsGFP70 = replicategroup.loc[0,'ts_GFP70']
    tsGFP100 = replicategroup.loc[0,'ts_GFP100']
    if mScarlet: 
        tsScarlet70 = replicategroup.loc[0,'ts_mScarlet70']
        tsScarlet100 = replicategroup.loc[0,'ts_mScarlet100']

    #We only want a timeseries point for each datapoint
    tsOD600 = tsOD600[0:leastData]
    minsOD600 = [get_mins(time) for time in tsOD600]
    
    tsGFP70 = tsGFP70[0:leastData]
    minsGFP70 = [get_mins(time) for time in tsGFP70]
    
    tsGFP100 = tsGFP100[0:leastData]
    minsGFP100 = [get_mins(time) for time in tsGFP100]
    
    if mScarlet:
        tsScarlet70 = tsScarlet70[0:leastData]
        minsScarlet70 = [get_mins(time) for time in tsScarlet70]

        tsScarlet100 = tsScarlet100[0:leastData]
        minsScarlet100 = [get_mins(time) for time in tsScarlet100]

    #Convert data to np.arrays
    ODarray, GFP70array, GFP100array = np.array(ODarray), np.array(GFP70array), np.array(GFP100array)
    tsOD600, tsGFP70, tsGFP100 = np.array(tsOD600), np.array(tsGFP70), np.array(tsGFP100)
    
    if mScarlet:
        SCARLET70array, SCARLET100array = np.array(SCARLET70array), np.array(SCARLET100array)
        tsScarlet70, tsScarlet100 = np.array(tsScarlet70), np.array(tsScarlet100)
    
#     print('pre-subtraction',GFP70array)
#     print('blanks-dict',blanksDict['blankGFP70median'][0:leastData])
    
    #background subtraction using the blanksDict, which has data on wells containing just media
    ODarray = ODarray-blanksDict['blankOD600median'][0:leastData]
    GFP70array = GFP70array-blanksDict['blankGFP70median'][0:leastData]
    GFP100array = GFP100array-blanksDict['blankGFP100median'][0:leastData]
    if mScarlet:
        SCARLET70array = SCARLET70array-blanksDict['blankSCAR70median'][0:leastData]
        SCARLET100array = SCARLET100array-blanksDict['blankSCAR100median'][0:leastData]
    
#     print('post-subtraction',GFP70array)
    
    GFP70arrayNormalized = GFP70array/ODarray
    
    if mScarlet:
        SCARLET100arrayNormalized = SCARLET100array/ODarray
    
    #Make sure replicates and associated timetraces have same number of datapoints
    if mScarlet:
        assert(all(array.shape==ODarray[0].shape for array in [ODarray[0],GFP70array[0],GFP100array[0],SCARLET70array[0],SCARLET100array[0],tsOD600,tsGFP70,tsGFP100,tsScarlet70,tsScarlet100]))
    else:
        assert(all(array.shape==ODarray[0].shape for array in [ODarray[0],GFP70array[0],GFP100array[0],tsOD600,tsGFP70,tsGFP100]))

        
    #Calculate mean and stdev for the data
    ODarrayMean = np.mean(ODarray, axis=0)
    GFP70arrayMean = np.mean(GFP70array, axis=0)
    GFP70arrayNormalizedMean = GFP70arrayMean/ODarrayMean
    
    if mScarlet:
        SCARLET70arrayMean = np.mean(SCARLET70array, axis=0)
        SCARLET100arrayMean = np.mean(SCARLET100array, axis=0)
        SCARLET100arrayNormalizedMean = SCARLET100arrayMean/ODarrayMean
    
    ODarraySTD = np.std(ODarray, axis=0)
    GFP70arraySTD = np.std(GFP70array, axis=0)
    GFP70arrayNormalizedSTD = np.std(GFP70arrayNormalized, axis=0)
    
    if mScarlet:
        SCARLET70arraySTD = np.std(SCARLET70array, axis=0)
        SCARLET100arraySTD = np.std(SCARLET100array, axis=0)
        SCARLET100arrayNormalizedSTD = np.std(SCARLET100arrayNormalized, axis=0)
    
    ODarraySE = stats.sem(ODarray, axis=0)
    GFP70arraySE = stats.sem(GFP70array, axis=0)
    GFP70arrayNormalizedSE = stats.sem(GFP70arrayNormalized, axis=0)
    
    if mScarlet:
        SCARLET70arraySE = stats.sem(SCARLET70array, axis=0)
        SCARLET100arraySE = stats.sem(SCARLET100array, axis=0)
        SCARLET100arrayNormalizedSE = stats.sem(SCARLET100arrayNormalized, axis=0)
    
    #Update the dictionary for that strain for induced replicates
    if mScarlet:
        replicateDict.update({'OD600':ODarray, 'GFP70':GFP70array, 'GFP70/OD600':GFP70arrayNormalized,'GFP100':GFP100array, 'SCAR70':SCARLET70array, 'SCAR100':SCARLET100array, 'SCAR100/OD600':SCARLET100arrayNormalized,
                            'OD600ts':tsOD600,'GFP70ts':tsGFP70,'GFP70/OD600ts':tsGFP70,'GFP100ts':tsGFP100,'SCAR70ts':tsScarlet70, 'SCAR100ts':tsScarlet100, 'SCAR100/OD600ts':tsScarlet100,
                            'OD600mins':minsOD600, 'GFP70mins':minsGFP70, 'GFP70/OD600mins':minsGFP70,'GFP100mins':minsGFP100,'SCAR70mins':minsScarlet70, 'SCAR100mins':minsScarlet100, 'SCAR100/OD600mins':minsScarlet100,
                            'OD600mean':ODarrayMean, 'GFP70mean':GFP70arrayMean, 'GFP70/OD600mean':GFP70arrayNormalizedMean, 'SCAR70mean':SCARLET70arrayMean, 'SCAR100mean':SCARLET100arrayMean, 'SCAR100/OD600mean':SCARLET100arrayNormalizedMean,
                            'OD600std':ODarraySTD, 'GFP70std':GFP70arraySTD, 'GFP70/OD600std':GFP70arrayNormalizedSTD, 'SCAR70std':SCARLET70arraySTD, 'SCAR100std':SCARLET100arraySTD, 'SCAR100/OD600std':SCARLET100arrayNormalizedSTD,
                            'OD600sem':ODarraySE, 'GFP70sem':GFP70arraySE, 'GFP70/OD600sem':GFP70arrayNormalizedSE, 'SCAR70sem':SCARLET70arraySE, 'SCAR100sem':SCARLET100arraySE, 'SCAR100/OD600sem':SCARLET100arrayNormalizedSE,
                            'strain':strain, 'num_datapoints':leastData, 'inducer_status':induction,  'Background':replicategroup.loc[0,'Background'],
                            'binding_sites':strToPrDF.loc[strain,'binding_sites'],
                            'parental_strain':strToPrDF.loc[strain,'parental_strain'],
                            'promoter_backbone':strToPrDF.loc[strain,'promoter_backbone']})
    else:
        replicateDict.update({'OD600':ODarray, 'GFP70':GFP70array, 'GFP70/OD600':GFP70arrayNormalized,'GFP100':GFP100array,
                            'OD600ts':tsOD600,'GFP70ts':tsGFP70,'GFP70/OD600ts':tsGFP70,'GFP100ts':tsGFP100,
                            'OD600mins':minsOD600, 'GFP70mins':minsGFP70, 'GFP70/OD600mins':minsGFP70,'GFP100mins':minsGFP100,
                            'OD600mean':ODarrayMean, 'GFP70mean':GFP70arrayMean, 'GFP70/OD600mean':GFP70arrayNormalizedMean, 
                            'OD600std':ODarraySTD, 'GFP70std':GFP70arraySTD, 'GFP70/OD600std':GFP70arrayNormalizedSTD, 
                            'OD600sem':ODarraySE, 'GFP70sem':GFP70arraySE, 'GFP70/OD600sem':GFP70arrayNormalizedSE, 
                            'strain':strain, 'num_datapoints':leastData, 'inducer_status':induction,  'Background':background,
                            'binding_sites':strToPrDF.loc[strain,'binding_sites'],
                            'parental_strain':strToPrDF.loc[strain,'parental_strain'],
                            'promoter_backbone':strToPrDF.loc[strain,'promoter_backbone']})
        
        
    return replicateDict
    
```

#### Processing methods

```python
def df_to_plotting_dict_Pho4(fdf, blanksDict):
    """
    This method takes the Zif268 platereader data dataframe and generates list of dictionaries
    (for the induced, and uninduced cases) that can be used for plotting (contain the mean signal, st.dev).
    """
    listInduced = []
    listNotInduced = []
    
    for info, strgrp, in fdf.groupby(["Background","Strain"]):
        strain = info[1]
        
        #Plan is to create a a list of dictionaries; one dictionary for each curve/dataset in the subsequent plots
        
        inducedReplicates = strgrp[strgrp.loc[:,'Media'] == 'Pi_free'].reset_index() # Currently assumes everything other than URA is induced
        dictInduced = process_replicates_Pho4(strain, inducedReplicates, blanksDict) #remove nan values, equalize num data points. Calculate  mean, stdev. Pack into dict.
        
        listInduced.append(dictInduced)
        
        notInducedReplicates = strgrp[strgrp.loc[:,'Media']=='Pi_rich'].reset_index()
        dictNotInduced = process_replicates_Pho4(strain, notInducedReplicates, blanksDict) #remove nan values, equalize num data points. Calculate  mean, stdev. Pack into dict.
                
        listNotInduced.append(dictNotInduced)
        
    return listInduced, listNotInduced
```

```python
def process_replicates_Pho4(strain, replicategroup, blanksDict):
    """
    This method takes as input a groupby of replicates from the Zif268 DataFrame. It processes 
    this data to remove nan values and equalize the number of data points. It calculates the 
    mean, and StDev. Then it packages everything into a dictionary, which it returns.
    """
    replicateDict = {}
    
    ODarray = []
    GFP70array = []
    GFP100array = []
    
    if replicategroup.loc[0,'Media'] == 'Pi_rich':
        inducer_status = 'uninduced'
    else:
        inducer_status = 'induced'
    
    #Combine all of the replicates into a 2D numpy array, remove nan values
    for i, replicate in replicategroup.iterrows():            
                
        ODarray.append([float(value) for value in replicate.loc['OD600'] if str(value) not in ['nan','OVRFLW']])
        GFP70array.append([float(value) for value in replicate.loc['GFP70'] if str(value) not in ['nan','OVRFLW']])
        
        if strain != 'sML051':
            GFP100array.append([float(value) for value in replicate.loc['GFP100'] if str(value) not in ['nan','OVRFLW']])
    
    if strain != 'sML051':
        leastData = min(min([len(array) for array in ODarray]), min([len(array) for array in GFP70array]), min([len(array) for array in GFP100array]))
    else:
        leastData = min(min([len(array) for array in ODarray]), min([len(array) for array in GFP70array]))

    
    ODarray = [array[0:leastData] for array in ODarray]
    GFP70array = [array[0:leastData] for array in GFP70array]
    if strain != 'sML051':
        GFP100array = [array[0:leastData] for array in GFP100array]

    #Get the timeseries' values
    tsOD600 = replicategroup.loc[0,'ts_OD600']
    tsGFP70 = replicategroup.loc[0,'ts_GFP70']
    if strain != 'sML051':
        tsGFP100 = replicategroup.loc[0,'ts_GFP100']

    #We only want a timeseries point for each datapoint
    tsOD600 = tsOD600[0:leastData]
    minsOD600 = [get_mins(time) for time in tsOD600]
    
    tsGFP70 = tsGFP70[0:leastData]
    minsGFP70 = [get_mins(time) for time in tsGFP70]
    
    if strain != 'sML051':
        tsGFP100 = tsGFP100[0:leastData]
        minsGFP100 = [get_mins(time) for time in tsGFP100]
    
    #Convert data to np.arrays
    ODarray, GFP70array, GFP100array = np.array(ODarray), np.array(GFP70array), np.array(GFP100array)
    tsOD600, tsGFP70  = np.array(tsOD600), np.array(tsGFP70)
    
    if strain != 'sML051':
        tsGFP100 = np.array(tsGFP100)
        
    #background subtraction using the blanksDict, which has data on wells containing just media
    ODarray = ODarray-blanksDict['blankOD600median'][0:leastData]
    GFP70array = GFP70array-blanksDict['blankGFP70median'][0:leastData]
    
    if strain != 'sML051':
        GFP100array = GFP100array-blanksDict['blankGFP100median'][0:leastData]

    
    GFP70arrayNormalized = GFP70array/ODarray
    if strain != 'sML051':
        GFP100arrayNormalized = GFP100array/ODarray
    
    #Make sure replicates and associated timetraces have same number of datapoints    
    if strain != 'sML051':
        assert(all(array.shape==ODarray[0].shape for array in [ODarray[0],GFP70array[0],GFP100array[0],tsOD600,tsGFP70,tsGFP100])) 
    else:
        assert(all(array.shape==ODarray[0].shape for array in [ODarray[0],GFP70array[0],tsOD600,tsGFP70]))
    
        #Calculate mean and stdev for the data
    ODarrayMean = np.mean(ODarray, axis=0)
    GFP70arrayMean = np.mean(GFP70array, axis=0)
    GFP70arrayNormalizedMean = GFP70arrayMean/ODarrayMean
    
    if strain != 'sML051':
        GFP100arrayMean = np.mean(GFP100array, axis=0)
        GFP100arrayNormalizedMean = GFP100arrayMean/ODarrayMean
        
    ODarraySTD = np.std(ODarray, axis=0)
    GFP70arraySTD = np.std(GFP70array, axis=0)
    GFP70arrayNormalizedSTD = np.std(GFP70arrayNormalized, axis=0)
    
    if strain != 'sML051':
        GFP100arraySTD = np.std(GFP100array, axis=0) 
        GFP100arrayNormalizedSTD = np.std(GFP100arrayNormalized, axis=0)
    
    ODarraySE = stats.sem(ODarray, axis=0)
    GFP70arraySE = stats.sem(GFP70array, axis=0)
    GFP70arrayNormalizedSE = stats.sem(GFP70arrayNormalized, axis=0)
    
    if strain != 'sML051':
        GFP100arraySE = stats.sem(GFP100array, axis=0)
        GFP100arrayNormalizedSE = stats.sem(GFP100arrayNormalized, axis=0)

    #Update the dictionary for that strain for induced replicates
    if strain != 'sML051':
        replicateDict.update({'OD600':ODarray, 'GFP70':GFP70array, 'GFP70/OD600':GFP70arrayNormalized, 'GFP100':GFP100array, 'GFP100/OD600':GFP100arrayNormalized,
                        'OD600ts':tsOD600,'GFP70ts':tsGFP70,'GFP70/OD600ts':tsGFP70, 'GFP100ts':tsGFP100, 'GFP100/OD600ts':tsGFP100,
                        'OD600mins':minsOD600, 'GFP70mins':minsGFP70, 'GFP70/OD600mins':minsGFP70, 'GFP100mins':minsGFP100, 'GFP100/OD600mins':minsGFP100,
                        'OD600mean':ODarrayMean, 'GFP70mean':GFP70arrayMean, 'GFP70/OD600mean':GFP70arrayNormalizedMean,'GFP100mean':GFP100arrayMean, 'GFP100/OD600mean':GFP100arrayNormalizedMean,
                        'OD600std':ODarraySTD, 'GFP70std':GFP70arraySTD, 'GFP70/OD600std':GFP70arrayNormalizedSTD, 'GFP100std':GFP100arraySTD, 'GFP100/OD600std':GFP100arrayNormalizedSTD,
                        'OD600sem':ODarraySE, 'GFP70sem':GFP70arraySE, 'GFP70/OD600sem':GFP70arrayNormalizedSE, 'GFP100sem':GFP100arraySE, 'GFP100/OD600sem':GFP100arrayNormalizedSE,
                        'strain':strain, 'Media':replicategroup.loc[0,'Media'], 'Background':replicategroup.loc[0,'Background'],
                        'num_datapoints':leastData, 'inducer_status':inducer_status,
                        'binding_sites':strToPrDF.loc[strain,'binding_sites'],
                        'promoter_backbone':strToPrDF.loc[strain,'promoter_backbone']})
    
    else:
        replicateDict.update({'OD600':ODarray, 'GFP70':GFP70array, 'GFP70/OD600':GFP70arrayNormalized,
                        'OD600ts':tsOD600,'GFP70ts':tsGFP70,'GFP70/OD600ts':tsGFP70,
                        'OD600mins':minsOD600, 'GFP70mins':minsGFP70, 'GFP70/OD600mins':minsGFP70,
                        'OD600mean':ODarrayMean, 'GFP70mean':GFP70arrayMean, 'GFP70/OD600mean':GFP70arrayNormalizedMean,
                        'OD600std':ODarraySTD, 'GFP70std':GFP70arraySTD, 'GFP70/OD600std':GFP70arrayNormalizedSTD,
                        'OD600sem':ODarraySE, 'GFP70sem':GFP70arraySE, 'GFP70/OD600sem':GFP70arrayNormalizedSE,
                        'strain':strain, 'Media':replicategroup.loc[0,'Media'], 'Background':replicategroup.loc[0,'Background'],
                        'num_datapoints':leastData, 'inducer_status':inducer_status,
                        'binding_sites':strToPrDF.loc[strain,'binding_sites'],
                        'promoter_backbone':strToPrDF.loc[strain,'promoter_backbone']})
    
    
    return replicateDict
```
